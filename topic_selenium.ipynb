```python
import pandas as pd

df = pd.read_csv('2-23_to_4-05.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
#df.head()
```

```python
df = df.drop(['Year'], axis=1)
df.head()
```

```python
df_march = df[df.Month == 3]
df_march
```

```python
df_march = df_march.drop(['Date', 'Day'], axis = 1)
df_march
```

```python
df_march = df_march.set_index('Writer')
df_march
```

```python
df_march = df_march.reset_index()
df_march
```

```python
df_march_akimoto = df_march[df_march.Writer == '아키모토마나츠']
df_march_hinata = df_march[df_march.Writer == 'HINATA46']
df_march_jo = df_march[df_march.Writer == '조내성.']
df_march_min = df_march[df_march.Writer == '민식']
```

```python
import pyLDAvis.sklearn
from sklearn.decomposition import LatentDirichletAllocation
import math
from soynlp.word import WordExtractor
from soynlp.tokenizer import LTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer

titles_akimoto = df_march_akimoto['Title'].values

word_ext = WordExtractor()
word_ext.train(titles_akimoto)
scores = word_ext.word_scores()
score_dict = {key: scores[key].cohesion_forward * math.exp(scores[key].right_branching_entropy) for key in scores}

lto = LTokenizer(scores=score_dict)
print(lto.tokenize(titles_akimoto[1]))



tfidf = TfidfVectorizer(ngram_range=(1, 1), min_df=3,
                        max_df=0.9, tokenizer=lto.tokenize, token_pattern=None)
topics_akimoto = tfidf.fit_transform(titles_akimoto)


lda = LatentDirichletAllocation(
    n_components=4, random_state=123, learning_method='batch')
topics_akimoto_finale = lda.fit_transform(topics_akimoto)

n_top_words = 10

feature_names = tfidf.get_feature_names()

for topic_idx, topic in enumerate(lda.components_):
    print("토픽 %d:" % (topic_idx + 1))
    print(" ".join([feature_names[i]
          for i in topic.argsort()[:-n_top_words - 1:-1]]))

pyLDAvis.enable_notebook()

vis = pyLDAvis.sklearn.prepare(lda, topics_akimoto, tfidf)
pyLDAvis.display(vis)
```
